```python
from docx import Document
import pandas as pd
from docx.enum.text import WD_PARAGRAPH_ALIGNMENT
from docx.oxml.ns import qn
from docx.oxml import OxmlElement
```

```python
#Preprarar dados
df = pd.read_csv('./disciplinas.csv', sep=',')
df.head()
```

```python

for linha in df.index:
    documento = Document('./PROGRAMA_DE_DISCIPLINA.docx')

    Disciplina = df.loc[linha, "Disciplina"]
    Sigla = df.loc[linha, "Sigla"]
    Ementa = df.loc[linha, "Ementa"]
    Bibliografia = df.loc[linha, "Bibliografia"]
    Nivel = df.loc[linha, "Nivel"]
    Credito = df.loc[linha, "Credito"]
    Carga_horaria = df.loc[linha, "Carga_horaria"]
    Programa = df.loc[linha, "Programa"]
    Tipo = df.loc[linha, "Tipo"]

    referencias = {
        "DDD": Disciplina,
        "ZZZ": Sigla,
        "XXX": Ementa,
        "BBB": Bibliografia,
        "NNN": Nivel,
        "YYY": str(Credito),
        "WWW": str(Carga_horaria), 
        "LLL": str(Tipo), 
        "PPP": Programa, 
    }
    for tabela in documento.tables:
        for linha in tabela.rows:
            for celula in linha.cells:

                if celula.text == "IDENTIFICAÇÃO":
                # Centralizar o texto
                    celula.text = celula.text.replace('IDENTIFICAÇÃO', '\t\t\t\t\tIDENTIFICAÇÃO')
                if celula.text == "PROGRAMA DE DISCIPLINA":
                # Centralizar o texto
                    celula.text = celula.text.replace('PROGRAMA DE DISCIPLINA', '\t\t\t\tPROGRAMA DE DISCIPLINA')
                  #  for paragrafo in celula.paragraphs:
                   #     paragrafo.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER
                    #    # Colocar o texto em negrito
                     #   for run in paragrafo.runs:
                      #      run.bold = True
                #Preencher tabela
                for codigo in referencias:
                    valor = referencias[codigo]
                    celula.text = celula.text.replace(codigo, valor)
    documento.save(f"./doc/Programa de Disciplina - {Disciplina}.docx")
```
